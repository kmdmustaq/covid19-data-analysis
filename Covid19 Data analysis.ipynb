# Data Cleaning

Lets consider not to drop any column or row where NaN is available insted NaN's will be replaced with zero for the further analysis since the data is very small if we drop this column's or row's it will get shrinked very much

```python
# Replace Nan with 0
data = data.replace(np.NaN, 0)
```

```python
# Percentage of NaN present in the dataset
round(100*(data.isnull().sum())/len(data.index), 2)
```
